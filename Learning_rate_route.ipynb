```python
import matplotlib.pyplot as plt
import numpy as np

def constant_learning_rate(lr, epochs):
    return [lr] * epochs

def step_decay_learning_rate(initial_lr, drop, epochs, drop_every):
    lr = initial_lr
    lrs = []
    for epoch in range(epochs):
        lr = initial_lr * (drop ** (epoch // drop_every))
        lrs.append(lr)
    return lrs

def exponential_decay_learning_rate(initial_lr, decay_rate, epochs):
    lrs = []
    for epoch in range(epochs):
        lr = initial_lr * (decay_rate ** epoch)
        lrs.append(lr)
    return lrs

def cosine_decay_learning_rate(initial_lr, epochs):
    lrs = []
    for epoch in range(epochs):
        lr = initial_lr * (0.5 * (1 + np.cos(np.pi * epoch / epochs)))
        lrs.append(lr)
    return lrs


epochs = 100
initial_lr = 0.1
drop = 0.5
drop_every = 20
decay_rate = 0.95


constant_lrs = constant_learning_rate(initial_lr, epochs)
step_decay_lrs = step_decay_learning_rate(initial_lr, drop, epochs, drop_every)
exponential_decay_lrs = exponential_decay_learning_rate(initial_lr, decay_rate, epochs)
cosine_decay_lrs = cosine_decay_learning_rate(initial_lr, epochs)


plt.figure(figsize=(12, 8))
plt.plot(constant_lrs, label='Constant Learning Rate')
plt.plot(step_decay_lrs, label='Step Decay Learning Rate')
plt.plot(exponential_decay_lrs, label='Exponential Decay Learning Rate')
plt.plot(cosine_decay_lrs, label='Cosine Decay Learning Rate')

plt.xlabel('Epochs')
plt.ylabel('Learning Rate')
plt.title('Learning Rate Schedules')
plt.legend()
plt.grid(True)
plt.show()
```

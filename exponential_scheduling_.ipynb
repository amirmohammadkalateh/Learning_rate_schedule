```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def exponential_scheduling(initial_value, decay_rate, num_points):
    """
    Generates values based on exponential scheduling.

    Args:
        initial_value (float): The initial value.
        decay_rate (float): The decay rate (between 0 and 1).
        num_points (int): The number of points to generate.

    Returns:
        ndarray: An array of values generated by exponential scheduling.
    """
    time = np.arange(num_points)
    values = initial_value * np.exp(-decay_rate * time)
    return time, values

def visualize_exponential_scheduling_2d(time, values, initial_value, decay_rate):
    """
    Visualizes exponential scheduling in 2D using matplotlib.

    Args:
        time (ndarray): Array of time points.
        values (ndarray): Array of scheduled values.
        initial_value (float): The initial value used for scheduling.
        decay_rate (float): The decay rate used for scheduling.
    """
    plt.figure(figsize=(10, 6))
    plt.plot(time, values, marker='o', linestyle='-')
    plt.title(f'Exponential Scheduling (2D)\nInitial Value: {initial_value}, Decay Rate: {decay_rate}')
    plt.xlabel('Time Step')
    plt.ylabel('Value')
    plt.grid(True)
    plt.show()

def visualize_exponential_scheduling_3d(time, values, initial_value, decay_rate):
    """
    Visualizes exponential scheduling in 3D using matplotlib.
    The z-axis represents another dimension (arbitrary for visualization).

    Args:
        time (ndarray): Array of time points.
        values (ndarray): Array of scheduled values.
        initial_value (float): The initial value used for scheduling.
        decay_rate (float): The decay rate used for scheduling.
    """
    fig = plt.figure(figsize=(12, 8))
    ax = fig.add_subplot(111, projection='3d')

    # Create a dummy z-dimension for visualization
    z = np.zeros_like(time)

    ax.plot(time, z, values, marker='o', linestyle='-')
    ax.set_xlabel('Time Step')
    ax.set_ylabel('Dimension 2 (Arbitrary)')
    ax.set_zlabel('Value')
    ax.set_title(f'Exponential Scheduling (3D)\nInitial Value: {initial_value}, Decay Rate: {decay_rate}')
    ax.view_init(elev=20, azim=-120)  # Adjust view angle for better perspective
    plt.show()

if __name__ == "__main__":
    initial_value = 1.0
    decay_rate = 0.1
    num_points = 50

    time, values = exponential_scheduling(initial_value, decay_rate, num_points)

    visualize_exponential_scheduling_2d(time, values, initial_value, decay_rate)
    visualize_exponential_scheduling_3d(time, values, initial_value, decay_rate)
```

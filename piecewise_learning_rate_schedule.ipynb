```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the piecewise learning rate schedule
def piecewise_lr_schedule(step):
    if step < 50:
        return 0.1
    elif step < 100:
        return 0.05
    elif step < 150:
        return 0.01
    else:
        return 0.005

# 2D Visualization
steps = np.arange(0, 200)
learning_rates = [piecewise_lr_schedule(s) for s in steps]

plt.figure(figsize=(10, 6))
plt.plot(steps, learning_rates, drawstyle='steps-post', color='blue')
plt.xlabel("Training Step (Epochs/Iterations)")
plt.ylabel("Learning Rate")
plt.title("Piecewise Learning Rate Schedule (2D)")
plt.grid(True)
plt.ylim(0)
plt.show()

# 3D Conceptual Visualization (Simplified)
# We'll visualize the learning rate as the height over a 2D parameter space
# This is a simplification as the parameter space is usually much higher dimensional.

# Define a simplified parameter space
param_x = np.linspace(-5, 5, 50)
param_y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(param_x, param_y)

# We'll associate different learning rates with conceptual "training stages"
# based on a proxy like the distance from the origin in the parameter space.
# This is just for visualization purposes and doesn't directly reflect the
# training process, but helps show the piecewise nature in 3D.

def conceptual_lr_surface(x, y):
    distance_squared = x**2 + y**2
    if distance_squared < 10:
        return 0.1
    elif distance_squared < 30:
        return 0.05
    elif distance_squared < 60:
        return 0.01
    else:
        return 0.005

Z_lr = np.array([[conceptual_lr_surface(x, y) for x in param_x] for y in param_y])

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z_lr, cmap='viridis', alpha=0.7)
ax.set_xlabel("Parameter 1")
ax.set_ylabel("Parameter 2")
ax.set_zlabel("Learning Rate")
ax.set_title("Conceptual Piecewise Learning Rate Surface (3D)")
fig.colorbar(surf, ax=ax, label='Learning Rate')
plt.show()

# Another way to visualize the 3D concept is to show how the learning rate
# changes over training steps (one dimension) and imagine the other two
# dimensions as the parameter space.

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Conceptual parameter dimensions (just for plotting)
param1 = np.zeros_like(steps)
param2 = np.zeros_like(steps)

# Plot the learning rate as the height over the "training step" plane
ax.plot(steps, param1, learning_rates, label='Learning Rate', color='blue')
ax.set_xlabel("Training Step")
ax.set_ylabel("Conceptual Parameter 1")
ax.set_zlabel("Learning Rate")
ax.set_title("Conceptual Piecewise Learning Rate over Training Steps (3D)")
ax.view_init(elev=30, azim=-120) # Adjust view angle
ax.legend()
plt.show()
```

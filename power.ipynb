```python

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def power_schedule(initial_lr, power, epoch):
    return initial_lr * (1 / (1 + epoch)) ** power

# 2D Plot
epochs = np.arange(0, 100)
initial_learning_rates = [0.1, 0.01, 0.001]
powers = [0.5, 1.0, 2.0]

plt.figure(figsize=(10, 5))
for lr in initial_learning_rates:
    for power in powers:
        lr_schedule = [power_schedule(lr, power, e) for e in epochs]
        plt.plot(epochs, lr_schedule, label=f'lr={lr}, power={power}')

plt.xlabel('Epoch')
plt.ylabel('Learning Rate')
plt.title('Power Scheduling - Learning Rate Decay')
plt.legend()
plt.grid(True)
plt.yscale('log')

# 3D Plot
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111, projection='3d')

power_range = np.linspace(0.1, 2.0, 50)
epoch_range = np.linspace(0, 100, 50)
P, E = np.meshgrid(power_range, epoch_range)
initial_lr = 0.1

LR = initial_lr * (1 / (1 + E)) ** P

surf = ax.plot_surface(P, E, LR, cmap='viridis')
ax.set_xlabel('Power')
ax.set_ylabel('Epoch')
ax.set_zlabel('Learning Rate')
ax.set_title('3D Power Scheduling Visualization')
fig.colorbar(surf)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def power_schedule(initial_lr, power, epoch):
    return initial_lr * (1 / (1 + epoch)) ** power

# Create data for contour plot
power_range = np.linspace(0.1, 2.0, 100)
epoch_range = np.linspace(0, 100, 100)
P, E = np.meshgrid(power_range, epoch_range)
initial_lr = 0.1

LR = initial_lr * (1 / (1 + E)) ** P

# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Regular contour plot
cs1 = ax1.contour(P, E, LR, levels=20)
ax1.clabel(cs1, inline=True, fontsize=8)
ax1.set_xlabel('Power')
ax1.set_ylabel('Epoch')
ax1.set_title('Contour Plot of Power Scheduling')

# Filled contour plot
cs2 = ax2.contourf(P, E, LR, levels=20, cmap='viridis')
ax2.set_xlabel('Power')
ax2.set_ylabel('Epoch')
ax2.set_title('Filled Contour Plot of Power Scheduling')
plt.colorbar(cs2, ax=ax2, label='Learning Rate')

plt.tight_layout()
plt.show()
```
